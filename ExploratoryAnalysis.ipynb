```r
library(ggplot2)
```

```r
#Loading the dataset
data <- read.csv('Loan.csv')
```

```r
head(data) #Viewing first 6 rows of data
```

```r
#Dealing with NA 
good <- complete.cases(data)
data <- data[good,]
```

```r
#Description of data
str(data)
```

```r
#Effect of education
qplot(Education, data = data,fill = Loan_Status)
```

```r
#Effect of marriage
qplot(Married, data = data,fill = Loan_Status)
```

```r
#Effect of Credit History
qplot(Credit_History, data = data,fill = Loan_Status)
```

```r
#Effect of Gender
qplot(Gender, data = data,fill = Loan_Status)
```

```r
#Effect of Property Area
qplot(Property_Area, data = data,fill = Loan_Status)
```

```r
#Distribution of Applicant's Income
qplot(ApplicantIncome,data=data,bins = 50)
qplot(Loan_Status,ApplicantIncome, data = data,geom = 'boxplot')
```

```r
qplot(ApplicantIncome,CoapplicantIncome,data=data,facets=.~Loan_Status)
```

```r
qplot(ApplicantIncome,LoanAmount,data=data,color=Loan_Status) + geom_smooth(method = 'lm',se = 0)
```

```r
#Distriution of Loan Amount
qplot(LoanAmount,data=data,bins = 50)
```
